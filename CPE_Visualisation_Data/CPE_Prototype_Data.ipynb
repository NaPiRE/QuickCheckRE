```python
# Import whole data from NaPiRe
cause_df = pd.read_csv(r'.\data\cause_lfdn.csv', encoding='latin1', delimiter=';', thousands=',')
effect_df = pd.read_csv(r'.\data\effect_lfdn.csv', encoding='latin1', delimiter=';', thousands=',')
```

```python
cause_df.head()
```

```python
cause_df = cause_df.rename(
    columns={'Source Id': 'source', 'Target Id': 'target', 'tag': 'label'})
condition = cause_df['target'] == 2.0
cause_df_2 = cause_df[condition]
cause_df_2
```

```python
row_data = {}
cause_data = []
row_percantage = {}

for index, row in cause_df_2.iterrows():
    row_percantage[row.source] = 0

for index, row in cause_df_2.iterrows():
    if row_percantage[row.source] > 0:
        row_percantage[row.source] += 1
    else:
        row_percantage[index] = 0
        row_data['id'] = 'l'+str(index)
        row_data['source'] = str(row.source).removesuffix('.0')
        row_data['target'] = str(row.target).removesuffix('.0')
        row_percantage[row.source] += 1
        cause_data.append(row_data)
        row_data = {}
cause_data
```

```python
node_data = []
node_rows = {}

# First element manually
node_rows['id'] = '2'
node_rows['label'] = 'Problem'
node_rows['data'] = {'nodeColor': '#004696'}
node_data.append(node_rows)
node_rows = {}

for index, row in cause_df_2.iterrows():
    node_rows['id'] = str(row.source).removesuffix('.0')
    node_rows['label'] = str(row.label)
    node_rows['data'] = {'nodeColor': '#004696'}
    node_data.append(node_rows)
    node_rows = {}


print(node_data)
```
