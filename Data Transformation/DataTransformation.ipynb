```python
jsonFile = open('./data/Quick-Check_Dump_20_06_2022.json')
surveyDump = js.load(jsonFile)

questionText = []
questionIds = []
questionTypes = []
questionChoices = []

for question in surveyDump['questions']:
    questionText.append(question['questionText']['en'])
    questionIds.append(question['id'])
    questionTypes.append(question['questionType'])
    if question['questionType'] != 'cause-problem-effect' and question['questionType'] != 'text':
        questionChoices.append(question['questionParameters'].get('choices'))
    elif question['questionType'] == 'cause-problem-effect':
        questionChoices.append([question['questionParameters'].get('problemChoices'), question['questionParameters'].get(
            'causeChoices'), question['questionParameters'].get('effectChoices')])
    else:
        questionChoices.append([])

questionDF = pd.DataFrame()
questionDF["question_id"] = pd.Series([], dtype=object)

for index, questionId in enumerate(questionIds):
    questionDF.loc[index, 'question_id'] = int(questionId)
    questionDF.loc[index, 'question_text'] = questionText[index]
    questionDF.loc[index, 'question_type'] = questionTypes[index]
    if(questionTypes[index] != 'text' and questionTypes[index] != 'cause-problem-effect'):
        for choiceIndex, choice in enumerate(questionChoices[index]):
            questionDF.loc[index, 'answer_' +
                           str(choiceIndex)] = choice['en']
    elif(questionTypes[index] == 'cause-problem-effect'):
        answerCounter = 0
        for problem in questionChoices[index][0]:
            questionDF.loc[index, 'answer_' +
                           str(answerCounter)] = 'PROBLEM_' + problem.get('en')
            answerCounter += 1
        for cause in questionChoices[index][1]:
            questionDF.loc[index, 'answer_' +
                           str(answerCounter)] = 'CAUSE_' + cause.get('en')
            answerCounter += 1
        for effect in questionChoices[index][2]:
            questionDF.loc[index, 'answer_' +
                           str(answerCounter)] = 'EFFECT_' + effect.get('en')
            answerCounter += 1


questions = questionDF.replace(np.nan, '')
questions.to_csv('./data/exports/questions.csv', sep=";")
questions
```

# Auxiliary functions

```python
def convertNapireToCce(questionsDF, toConvertDF, questionIndex):
    '''
    Converts values from the NaPiRE data format into the answer.csv fromat of
    the CCE tool importing script.

            Parameters:
                    questionsDF (df):    The data from question.csv
                    toConvertDF (df):    The NaPiRE data
                    questionIndex (int): Index of the question to be converted
                                         inside the question.csv

            Returns:
                    converted (df):      The resulting data frame in CCE
                                         format
    '''
    respondentAnswers = []
    converted = []
    for index, row in toConvertDF.iterrows():
        for label, content in questions.items():
            if label.startswith('answer_') and not (content[questionIndex] == ''):
                if row[1].replace(" ", "") == content[questionIndex].replace(" ", ""):
                    respondentAnswers.append((row[0], 'TRUE'))
                else:
                    respondentAnswers.append((row[0], 'FALSE'))
        converted.append(respondentAnswers)
        respondentAnswers = []
    return converted


def createQuestionColumns(questionPrefix, length):
    # Create resulting list with initial lfdn column
    questionColumn = ['lfdn']
    for index in np.arange(length):
        questionColumn.append(questionPrefix+(str(index)))
    return questionColumn

# return DataFrame with in CCE Tool Question.csv format


def createQuestionDF(questionColumn, convertedList):
    resultDF = pd.DataFrame(columns=questionColumn)
    for index, element in enumerate(convertedList):
        resultDF.loc[index] = element[0][0]
        for innerIndex, value in enumerate(element):
            resultDF.iloc[index, innerIndex+1] = value[1]
    return resultDF


def joinAllDataFrames(dataFrameList):
    result = dataFrameList[0]
    for index, elem in enumerate(dataFrameList):
        if index > 0:
            result = result.merge(elem)
    return result

# Converts and appends converted NaPiRE data to cce tool


def convertAndAppend(dfToConvert, questionListID, prefix):
    convertedDF = convertNapireToCce(questions, dfToConvert, questionListID)
    questionColumn = createQuestionColumns(prefix, len(convertedDF[0]))
    resultDF = createQuestionDF(questionColumn, convertedDF)
    resultList.append(resultDF)
    return resultDF
```
