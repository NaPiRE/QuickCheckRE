# Auxiliary functions

```python
def convertNapireToCce(questionsDF,toConvertDF,questionIndex):
    '''
    Converts values from the NaPiRE data format into the answer.csv fromat of
    the CCE tool importing script.

            Parameters:
                    questionsDF (df):    The data from question.csv
                    toConvertDF (df):    The NaPiRE data
                    questionIndex (int): Index of the question to be converted
                                         inside the question.csv

            Returns:
                    converted (df):      The resulting data frame in CCE
                                         format
    '''
    respondentAnswers = []
    converted = []
    for index,row in toConvertDF.iterrows():
        for label, content in questions.items():
            if label.startswith('answer_') and not (content[questionIndex] == ''):
                if row[1] == content[questionIndex]:
                    respondentAnswers.append((row[0],'TRUE'))
                else:
                    respondentAnswers.append((row[0],'FALSE'))
        converted.append(respondentAnswers)
        respondentAnswers=[]
    return converted

def createQuestionColumns(questionPrefix,length):
    # Create resulting list with initial lfdn column
    questionColumn = ['lfdn']
    for index in np.arange(length):
        questionColumn.append(questionPrefix+(str(index)))
    return questionColumn

# return DataFrame with in CCE Tool Question.csv format
def createQuestionDF(questionColumn,convertedList):
    resultDF = pd.DataFrame(columns=questionColumn)
    for index, element in enumerate(convertedList):
        resultDF.loc[index] = element[0][0]
        for innerIndex, value in enumerate(element):
            resultDF.iloc[index,innerIndex+1] = value[1]
    return resultDF
        
def joinAllDataFrames(dataFrameList):
    result = dataFrameList[0]
    for index, elem in enumerate(dataFrameList):
        if index > 0:
            result = result.merge(elem)
    return result

# Converts and appends converted NaPiRE data to cce tool
def convertAndAppend(dfToConvert,questionListID,prefix):
        convertedDF = convertNapireToCce(questions,dfToConvert,questionListID)
        questionColumn = createQuestionColumns(prefix,len(convertedDF[0]))
        resultDF = createQuestionDF(questionColumn,convertedDF)
        resultList.append(resultDF)
        return resultDF
```
